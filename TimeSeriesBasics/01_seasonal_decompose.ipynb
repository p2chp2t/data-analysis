```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./drive/MyDrive/Colab Notebooks/Colab-TimeSeriesBasics/quebec_real_estate.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head(10)
```

```python
data = df[['Montreal_Median_Price']].copy()
data.dropna(inplace=True)
data.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result1 = seasonal_decompose(data, model='multiplicative')
result1.plot()
```

```python
result2 = seasonal_decompose(data, model='additive')
result2.plot()
```
