```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('./drive/MyDrive/Colab Notebooks/Colab-TimeSeriesBasics/daily-min-temperatures.csv', header=0, index_col=0)
df.plot()
```

```python
lag=100
##### ACF #####
sm.graphics.tsa.plot_acf(df, lags=lag)
plt.show()
```

```python
##### PACF #####
sm.graphics.tsa.plot_pacf(df, lags=lag)
plt.show()
```
